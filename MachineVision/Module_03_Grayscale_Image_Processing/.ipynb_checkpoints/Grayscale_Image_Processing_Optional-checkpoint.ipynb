# Introduction

In this optional notebook you will learn more gray-scale processing techniques. The first one, called histogram equalization, is used to get more contrast in your image. The other one, $\gamma-$correction, is actually used in practically all screens and camera's, not only for gray-scale images. This is because camera's and the human eye register changes in light intensity different. 

A camera registers a twice as high signal if the intensity of the light is twice as high. The human eye doesn't work like that. In a very light environment the doubling of the intensity doesn't do much, while it does in a dark environment. This enables us to see contrasts in the dark.

We start with loading the packages we will need and define two functions with which we can easily plot images and their histograms next to each other. So, please evaluate the following two cells.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# A function to show a gray-scale image with its histogram side-by-side
def show_im_and_hist(originalImage, hist):
    plt.figure(figsize=(15,8))                              #create figure and set figure size
    plt.subplot(121)
    plt.imshow(originalImage, cmap='gray')
    plt.xticks([]), plt.yticks([])
    plt.title('Image')

    plt.subplot(122)
    plt.bar(list(range(len(hist))),hist[:,0])                # Plot the histogram as a bar plot
    plt.title('histogram')
    plt.xlabel("Intensity"), plt.ylabel("count")
    plt.show()
    
#Function to show one or multiple images
def show_images(images):
    figwidth = 20; figheight = figwidth * images[0][0].shape[0]/images[0][0].shape[1]
    plt.figure(figsize=(figwidth,figheight))
    cols = 2
    rows = len(images) // 2 + 1
    for i, image in enumerate(images):
        plt.subplot(rows,cols,i+1)
        plt.imshow(image[0], cmap='gray', vmin=0, vmax=255)
        plt.title(str(image[1]))
        plt.xticks([]), plt.yticks([])
    plt.show()
    
# Function to show an RGB image
def imshow_rgb(img_bgr):
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
```

## Histogram equalization
---

To illustrate the use of histogram equalization we load an image and histogram of an image in which the histogram is very unequal.

```python
img = cv2.imread('Data_Tutorial3/Kool.png',0)#load image
nr_of_bins = 256             # The number of bins in the histogram   
intensity_range = [0,256]    # The range of intensities captured in the histogram
hist = cv2.calcHist([img],[0],None,[nr_of_bins],intensity_range)  #apply histogram
show_im_and_hist(img,hist)
```

As you can see that the above histgrams are not very equalized. That is, the histogram is far from a uniform distribution. That is not very surprising, the image is drawn in black and white. __Histogram equalization__ is a method to distribute the intensities more equally. The figure below shows het idea behind histogram equalization.
<img src="Data_Tutorial3/histogram_equalization.png" />

Histogram equalization of an image can be performed by the function `cv2.equalizeHist`:

```python
imgCrowd = cv2.imread('Data_Tutorial3/Crowd.png',0) #load image
imgCrowd_eq = cv2.equalizeHist(imgCrowd)            # apply histogram equalization

# calculate the histograms of the original and equalized image
hist = cv2.calcHist([imgCrowd],[0],None,[256],[0,256])           #calculate histogram
hist_eq = cv2.calcHist([imgCrowd_eq],[0],None,[256],[0,256])     #calculate histogram

print("Original image")
show_im_and_hist(imgCrowd, hist)
print("Histogram-equalized image")
show_im_and_hist(imgCrowd_eq, hist_eq)
```

## Intensity transformation
---

The brightness of pixels in an image can be changed using __intensity transformations__. The transformation can be used to increase or decrease the contrast in certain intensity ranges. __Gamma correction__ is a popular way to transform the intensity. Based on the parameter $\gamma$, different transformation functions can easily be made. The image below shows the transformation functions for different values of $\gamma$.

<img src="Data_Tutorial3/gamma_transformations.png" width="400px">

```python
# Function to get a gamma intensity-transformation 
def get_gamma_transformation(gamma):
    gamma = 0 if gamma<0 else gamma  # Gamma needs to be greater or equal to 0.
    gamma_trans = [((i / 255.0) ** gamma) * 255  for i in np.arange(0, 256)]    
    return np.array(gamma_trans).astype("uint8")#change the pixelvalues to unint8

# Function to apply the gamma correction using look-up tables (LUT)
def gamma_correction(image, gamma):
    gamma_trans = get_gamma_transformation(gamma)
    return cv2.LUT(image, gamma_trans)
```

The code above creates two functions, one to get a gamma intensity-transformation for a given gamma and one to apply this transformation to the image.

Let's create a few transformation functions with different gamma values and see how they look.

```python
for i,gamma in enumerate((0.20, 1.0, 5.0)):
    gamma_trans = get_gamma_transformation(gamma)
    plt.plot(gamma_trans, color='C%d'%i, label='gamma = %1.2f'%gamma)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```
