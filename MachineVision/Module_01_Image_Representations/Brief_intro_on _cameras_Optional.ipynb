```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
# Set figure size so that images are large enough
plt.rcParams['figure.figsize'] = [20, 10]

def imshow_rgb(img_bgr):                            #function to display an image using RGB color format.                                                       #read images, plt.imshow builds a window with the image in it, plt.show 
    img_rgb = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
```

## Effects of camera and lens settings
---

### Shutter time
We will load and show a number of images that have been taken with a digital camera. The same scene is photographed with different shutter times (0.5ms, 2ms, 56ms and 220ms)

```python
# Create an array with the shutter times, so that we can load the correct image
shutter_times = ["0,0005", "0,002", "0,00556", "0,022"]
                   
for i,shutter_time in enumerate(shutter_times): # Using enumerate to get the shutter time and an iterator (0,1,2..)
    # Construct the image filename
    image_file = ('Data_Tutorial1/Tv%s AV4,0.JPG' % shutter_time)
    # Show the image in a 2x2 grid
    plt.subplot(2,2,i+1)
    img_bgr = cv2.imread(image_file)
    imshow_rgb(img_bgr)
    plt.title('Shutter time: %s ms' % shutter_time)
                           
plt.show()
```

### Lens aperture 
We will now keep the shutter time fixed and change the lens aperture (diaphragm): f/4.0, f/6.9, f/13, and f/22.

```python
# Create an array with the shutter times, so that we can load the correct image
apertures = ["4,0", "6,9", "13", "22"]
                   
for i,aperture in enumerate(apertures):
    # Construct the image filename
    image_file = ('Data_Tutorial1/TV0,004_AV%s.JPG' % aperture)
    # Show the image in a 2x2 grid
    plt.subplot(2,2,i+1)
    img_bgr = cv2.imread(image_file)
    imshow_rgb(img_bgr)
    plt.title('Lens aperture: f/%s' % aperture)
                           
plt.show()
```

### Depth of field
You noticed that both the shutter time and the lens aperture have an influence on the brightness of the image. We can use this to make sure to get the correct brightness of the image (not under-exposed and not over-exposed).

In the code below, two images are loaded. One with a short shutter time and wide aperture, and one with a longer shutter imte and narrow aperture.

__Exercise (depth of field):__
1. Run the code an describe what the influence of the lens aperture is on the __depth of field__ of the image.


```python
shutter_times = ["0,002", "0,05"]
apertures = ["5,6", "22"]

for i in range(0,2):
    image_file = ('Data_Tutorial1/TV%s_AV%s.JPG' % (shutter_times[i], apertures[i]))
    plt.subplot(1,2,i+1)
    img_bgr = cv2.imread(image_file)
    imshow_rgb(img_bgr)
    plt.title('Shutter time: %s, aperture: %s' % (shutter_times[i], apertures[i]))
plt.show()
```
